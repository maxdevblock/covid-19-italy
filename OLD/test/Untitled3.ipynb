```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
SE = np.linspace(0, 1, 1001)
SP = np.linspace(0, 1, 1001)
```

```python
def LRp(se, sp):
    lr = se / (1 - sp)
    #print(f"LR={lr[20]}")
    return lr

def LRn(se, sp):
    return (1 -se) / sp

def SP_from_LRp(se, lr):
    return 1 - se / lr

def LR_OR(se, sp, r=2):
    _se = 1 - (1 - se) ** r
    _sp = sp ** r
    #print(f"""SE({r})={_se[20]}  SP({r})={_sp[20]}""")
    return LRp(_se, _sp)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

LR = 100
SPx = SP_from_LRp(SE, LR)
#ax.plot(SE, [LR for _ in range(len(SE))])
_LR = LR

for i in range(10):
    _SP = SP_from_LRp(SE, _LR)
    _LR = LR_OR(SE, _SP, r=i+1)
    ax.plot(SE, _LR, label=f"{LR}({i})")

LR = 50
SPx = SP_from_LRp(SE, LR)
#ax.plot(SE, [LR for _ in range(len(SE))])
_LR = LR

for i in range(10):
    _SP = SP_from_LRp(SE, _LR)
    _LR = LR_OR(SE, _SP, r=i+1)
    ax.plot(SE, _LR, ls="--", label=f"{LR}({i})")

LR = 25
SPx = SP_from_LRp(SE, LR)
#ax.plot(SE, [LR for _ in range(len(SE))])
_LR = LR

for i in range(10):
    _SP = SP_from_LRp(SE, _LR)
    _LR = LR_OR(SE, _SP, r=i+1)
    ax.plot(SE, _LR, ls=":", label=f"{LR}({i})")

ax.legend(x=1.25)
```
