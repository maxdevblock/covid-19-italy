```python
import numpy as np
import pandas as pd
import scipy.stats as sps
import matplotlib.pyplot as plt
```

```python
a = np.array([.938, .9735, .9255, .82, 1, .969, .886, .841, .87, .91, .82, .95, .931, .931, .934, .9585])

b = np.array([.956, .99, 1, .97, .987, .994, .9063, .923, .955, .967, .96, 1, 1, .991, .977, .9935])
```

```python
x = np.linspace(min(a), max(a), 50)
f = sps.cauchy.fit(a)
y = sps.cauchy.pdf(x, *f)
plt.hist(pd.Series(a), density=True, alpha=.25, ec="b", bins=7)
plt.plot(x, y, lw=5)
plt.show();
```

```python
x = np.linspace(min(b), max(b), 50)
f = sps.skewnorm.fit(b)
y = sps.skewnorm.pdf(x, *f)
plt.hist(pd.Series(b), density=True, alpha=.25, ec="b", bins=7)
plt.plot(x, y, lw=5)
plt.show();
```
