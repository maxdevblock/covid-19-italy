```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
```

```python
# beta
plt.style.use('dark_background')
fig, ax = plt.subplots(1, 2, figsize=(10,4))
a = 2
b = 3
omega = (a-1)/(a+b-2)
kappa = a+b
mu = a / kappa
x = np.linspace(sps.beta.ppf(0, a=a, b=b), sps.beta.ppf(1, a=a, b=b), 100)
y = sps.beta.pdf(x, a=a, b=b)
ax[0].plot(x, y, lw=7)
ax[0].axvline(omega, c="w", ls=":")
ax[0].set_xlim(-0.1,1.1)
ax[0].axhline(0, c="w", alpha=.5)
ax[0].set_title("Beta distribution", fontsize=20)
ax[0].text(
    .7, 1.6,
    f"""a={a} b={b}
$\mu={mu}$ $\kappa={kappa}$
$\omega={omega:.2f}$""",
    ha="left", va="top", fontsize=15
)

ax[1].text(
    .1, .5,
    r"""
$a = \omega(\kappa - 2) + 1$
$a = \mu \kappa$

$b = (1 - \omega)(\kappa - 2) + 1$
$b = (1 - \mu) \kappa$

$\omega = (a-1) / (1+b-2)$
$\mu = a / (a + b)$
$\kappa = a + b$
    """,
    ha="left", va="center",
    fontsize=22
)
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[1].axis('off')
plt.tight_layout()
plt.savefig("beta")
plt.show();
```

```python
# gamma
plt.style.use('dark_background')
fig, ax = plt.subplots(1, 2, figsize=(10,4))
shape = 1.56
rate = .0312
mu = shape / rate
omega = (shape - 1) / rate
sigma = np.sqrt(shape) / rate

x = np.linspace(sps.gamma.ppf(1e-10, a=shape, scale=1/rate), sps.gamma.ppf(1-1e-10, a=shape, scale=1/rate), 1000)
y = sps.gamma.pdf(x, a=shape, scale=1/rate)
ax[0].plot(x, y, lw=7, c="C3")
ax[0].axvline(omega, c="w", ls=":")
ax[0].set_xlim(-10,150)
ax[0].axhline(0, c="w", alpha=.5)
ax[0].set_title("Gamma distribution", fontsize=20)
ax[0].text(
    70, .013,
    f"""s={shape} r={rate}
$\mu={mu:.0f}$ $\sigma={sigma:.0f}$
$\omega={omega:.0f}$""",
    ha="left", va="top",
    fontsize=15
)

ax[1].text(
    .1, .5,
    r"""$s = \mu^2 / \sigma^2$
$r = \mu / \sigma^2$

$s = 1 + \omega r$
$r = ( \omega + \sqrt{\omega^2 + 4\sigma^2} ) / 2\sigma^2 $

$\omega = (s-1)/r$
$\mu = s / r$
$\sigma = \sqrt{s} / r$""",
    ha="left", va="center",
    fontsize=22
)
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[1].axis('off')
plt.tight_layout()
plt.savefig("gamma")
plt.show();
```
