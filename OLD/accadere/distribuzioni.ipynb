# Discrete

```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
```

```python
def plot_dist(dist, y, *args):
    print(args)
    fig, ax = plt.subplots(1, 3, figsize=(7.5,2.5))
    fig.suptitle(f"DISTRIBUZIONE {dist.name}", fontsize=20)

    ax[0].set_xlabel("Density")
    ax[0].stem(dist.pmf(y, *args), use_line_collection=True)
    
    ax[1].set_xlabel("Cumulativ")
    ax[1].plot(dist.cdf(y, *args))
    
    ax[2].set_xlabel("Random Samples")
    ax[2].scatter(np.arange(10), dist.rvs(*args, size=10))
```

```python
plot_dist(
    sps.bernoulli,
    np.arange(2),
    .75
)
```

```python
plot_dist(
    sps.betabinom,
    np.arange(10),
    10, 3, 7
)
```

```python
plot_dist(
    sps.binom,
    np.arange(10),
    10, .7
)
```

```python
plot_dist(
    sps.boltzmann,
    np.arange(10),
    .25, 7
)
```

```python
plot_dist(
    sps.dlaplace,
    np.arange(10),
    .25
)
```

```python
plot_dist(
    sps.geom,
    np.arange(10),
    .3
)
```

```python
plot_dist(
    sps.hypergeom,
    np.arange(10),
    10, 5, 7
)
```

```python
plot_dist(
    sps.logser,
    np.arange(10),
    .75
)
```

```python
a = np.arange(8)[1:]
b = [a for _ in range(100)]
sps.multinomial.pmf(x=b, n=1, p=[i/sum(a) for i in a])
```

```python
plt.matshow(sps.multinomial.rvs(n=1, p=[i/sum(a) for i in a], size=100))
```

```python
plt.matshow(sps.multinomial.cov(n=1, p=[i/sum(a) for i in a]))
```

```python
plt.stem(sps.rv_discrete(values=(a, [i/sum(a) for i in a])).pmf(a))
```
