```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display, Markdown
```

```python
P_M = 0
P_S = 0
P_pos = 0
P_neg = 0
```

```python
P_M = 223000 / 60501878
P_M
```

```python
P_S = 1 - P_M
P_S
```

```python
P_pos_M = np.linspace(0, 1, 11)
P_pos_M
```

```python
P_neg_M = 1 - P_pos_M
P_neg_M
```

```python
P_pos_S = np.linspace(0, 1, 11)
P_pos_S
```

```python
P_neg_S = 1 - P_pos_S
P_neg_S
```

```python
P_pos_S * P_S
```

```python
P_pos = (P_pos_M * P_M) + (P_pos_S * P_S)[:, None]
P_pos_df = pd.DataFrame(P_pos, index=P_pos_S, columns=P_pos_M)
P_pos_df
```

```python
P_neg = (P_neg_M * P_M) + (P_neg_S * P_S)[:, None]
P_neg_df = pd.DataFrame(P_neg, index=P_neg_S, columns=P_neg_M)
P_neg_df
```

```python
for p_neg_s in P_neg_S:
    print(f"P(-|S) = {p_neg_s} -----------------------------")
    P_S_neg = (p_neg_s * P_S) / P_neg_df
    display(Markdown(P_S_neg.round(3).to_html()))
```
