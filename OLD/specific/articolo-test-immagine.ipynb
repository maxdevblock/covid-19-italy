```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import dates as mdates
```

```python
url_ita = 'https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-andamento-nazionale/dpc-covid19-ita-andamento-nazionale.csv'
ita = pd.read_csv(url_ita,
                     usecols=['data', 'tamponi', 'casi_testati'],
                     parse_dates=['data'],
                     index_col=['data'],
                     squeeze=True).sort_index()

df = ita.loc["2020-04-19 17:00:00":]
```

```python
with plt.xkcd():
    fig, ax = plt.subplots(figsize=(8, 8))

    ax.bar(df.index, df["casi_testati"], label="singoli casi testati")
    ax.bar(df.index, df["tamponi"]-df["casi_testati"], bottom=df["casi_testati"], label="tamponi eccedenti")

    ax.xaxis.set_major_locator(mdates.WeekdayLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))

    ax.legend(loc="upper left", fontsize=20)
    
    #ax.set_title("Casi testati rispetto ai tamponi effettuati")
plt.show();
```

```python
with plt.xkcd():
    fig, ax = plt.subplots(figsize=(8, 8))

    ax.plot(df.index, df["casi_testati"], label="singoli casi testati", lw=5)
    ax.fill_between(df.index, [0 for _ in range(len(df.index))], df["casi_testati"], alpha=.5)
    #ax.text(df.index[df.index.size // 2], 8e5, "singoli casi testati", ha="center", fontsize=30, rotation=20)
    
    ax.plot(df.index, df["tamponi"], label="tamponi eccedenti", lw=5)
    ax.fill_between(df.index, df["casi_testati"], df["tamponi"], alpha=.5)
    #ax.text(df.index[df.index.size // 2], 1.5e6, "tamponi eccedenti", ha="center", fontsize=30, rotation=20)

    ax.xaxis.set_major_locator(mdates.WeekdayLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
    #ax.set_yticks([])
    ax.legend(loc="upper left", fontsize=20)

    #ax.set_title("Casi testati rispetto ai tamponi effettuati")
plt.show();
```
